# 기초상품 추천 코드 구현해보기
주어진 데이터를 활용하여 유저 아이디와 원하는 제품 유형을 입력하면   
해당 유저와 유사도가 높은 유저의 평점 높은 제품을 추천해주는 함수 만들기

* 유저데이터 파일명 : user_data
* 제품데이터 파일명 : product_data

```python
import pandas as pd

#유저데이터 불러오기
u_data = pd.read_csv('user_data.csv')

#제품데이터 불러오기
p_data = pd.read_csv('product_data.csv', encoding='cp949')
```

```python
#u_data 확인
u_data
```

```python
#p_data 확인
p_data
```

```python
#product_id 기준으로 유저,제품 데이터 합쳐서 df에 저장하고 user_id를 기준으로 오름차순 정렬

df = u_data.merge(p_data, on='product_id', how='inner')
df = df.sort_values(by='user_id')

df
```

```python
#각 유저가 제품에 남긴 평점들로 이루어진 데이터 프레임(df_users) 만들기 인덱스:user_id, 컬럼명:product_id
df_users = pd.pivot_table(df, values='rating', index='user_id', columns='product_id')
df_users.fillna(0, inplace=True)
df_users.head()
```

```python
#유저간 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity
cos_matrix = cosine_similarity(df_users)
cos_matrix
```

```python
#위에서 계산한 cos_matrix를 데이터프레임으로 만듭니다. 
df_users_cosine= df_users_cosine = pd.DataFrame(cos_matrix, index=df_users.index, columns=df_users.index)
df_users_cosine.head()
```

## 입력한 아이디의 유저와 비슷한 유저가 만족한 제품 추천하는 함수 만들기
1. df_users_cosine에서 입력한 아이디와 유사도 높은 5명을 sim_users로 지정
2. df에서 user_id가 sim_users의 인덱스와 일치하는 값들을 sim_user_df에 할당
3. sim_user_df에서 입력한 product_type과 일치하는 값들을 지정하고 rating을 기준으로 내림차순한 결과를 products에 할당
4. products에서 rating 4점 이상인 값들만 다시 products에 할당
5. products에서 product_name이 중복인 것은 첫번째 값만 남김
6. products의 product_name을 데이터 프레임 result로 만듦 (컬럼명은 나와 비슷한 사용자가 만족한(입력한 제품타입)제품)
7. 함수 마지막에 print(result) 출력

```python
#반드시 skeleton 코드에 맞춰서 작성하지 않아도 됨. 단, 작성한 코드에 대해 어떤 알고리즘으로 결과가 나오게 했는지 설명할 수 있어야함. 

def user_based_recommend(user_id, product_type):
    # 입력한 아이디와 유사한 유저들 추출
    sim_users = df_users_cosine[user_id].sort_values(ascending=False)[1:6].index.tolist()

    sim_user_df = df[df['user_id'].isin(sim_users)]

    products = sim_user_df[sim_user_df['product_type'] == product_type].sort_values(by='rating', ascending=False)
    products = products[products['rating'] >= 4]
    products = products.drop_duplicates(subset=['product_name'], keep='first')
    
    result=pd.DataFrame(products['product_name'])
    result.reset_index(drop=True, inplace=True)
    result.columns = [f"나와 비슷한 사용자가 만족한 {product_type} 제품"]

    print(result)
```

```python
# 함수 사용 예시
user_based_recommend(1, '토너')
```

## +) 알레르기성 제품 제외 추천 시스템
알레르기가 있는 고객들을 위해 알레르기 성분이 포함되어 있는 제품을 필터링하는 추천 시스템을 구현해보았습니다

```python
def user_based_recommend2(user_id, product_type, allergic_component):
    # 입력한 아이디와 유사한 유저들 추출
    sim_users = df_users_cosine[user_id].sort_values(ascending=False)[1:6].index.tolist()

    sim_user_df = df[df['user_id'].isin(sim_users)]

    products = sim_user_df[sim_user_df['product_type'] == product_type].sort_values(by='rating', ascending=False)
    products = products[products['rating'] >= 4]

    # 알레르기 성분을 고려하여 제품 필터링
    products = products[~products['ingredient'].apply(lambda x: allergic_component in x)]

    products = products.drop_duplicates(subset=['product_name'], keep='first')
    
    result=pd.DataFrame(products['product_name'])
    result.reset_index(drop=True, inplace=True)
    result.columns = [f"나와 비슷한 사용자가 만족한 {product_type} 제품"]

    print(result)
```

```python
#알레르기 성분이 없는 경우 예시
user_based_recommend2(1, '토너', '없음')
```

```python
#알레르기 성분이 있는 경우 예시1
user_based_recommend2(1, '토너', '포도')
```

```python
#알레르기 성분이 있는 경우 예시2
user_based_recommend2(1, '토너', '쑥잎추출물')
```
